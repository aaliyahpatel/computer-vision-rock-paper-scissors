```python
import cv2
from keras.models import load_model
import numpy as np
model = load_model('keras_model.h5')
```

```python
img = cv2.imread("/Users/aaliyahpatel/Desktop/0.jpg")
```

```python
import matplotlib.pyplot as plt
img = img[:,:,::-1]
plt.imshow(img)


```

```python
resized_frame = cv2.resize(img, (224, 224), interpolation = cv2.INTER_AREA)
image_np = np.array(resized_frame)
normalized_image = (image_np.astype(np.float32) / 127.0) - 1
```

```python
normalized_image.shape
```

```python
normalized_image = normalized_image.reshape(1,224,224,3)
```

```python
pred = model.predict(normalized_image)
```
